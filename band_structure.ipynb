```python
from math import *
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
```

```python
bands = np.loadtxt('band_energies.dat').T
bands = bands - np.min(bands)
```

```python
plt.figure(figsize = (6.4, 3.2))

for i, band in enumerate(bands):
    plt.plot(band, linewidth = 3.0, label = f'Band {i + 1}')

plt.title(r'Al Band Structure', fontsize = 12)
plt.xlabel(r'k-path', fontsize = 12)
plt.ylabel(r'Energy (eV)', fontsize = 12)

ax = plt.gca()
ax.xaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator(4))
ax.yaxis.set_minor_locator(matplotlib.ticker.AutoMinorLocator(4))

plt.xticks([0, 10, 20, 25], [r'$\Gamma$', r'$X$', r'$\Gamma$', r'$L$'])
plt.yticks([0, 10, 20, 30, 40])
plt.tick_params(direction = 'in', which = 'both', labelsize = 12)
plt.legend(frameon = False, fontsize = 12, loc = 'center left', bbox_to_anchor = (1.05, 0.5))
plt.grid(zorder = 0, axis = 'x')
plt.tight_layout()
plt.savefig('Al_band_structure.png')
plt.savefig('Al_band_structure.pdf')
plt.show()
```
